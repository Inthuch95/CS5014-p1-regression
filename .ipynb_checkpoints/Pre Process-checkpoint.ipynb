```python
from datetime import datetime, timedelta
import numpy as np

def nsm(dt_obj):
    """Get the number of seconds until midnight."""
    tomorrow = dt_obj + timedelta(1)
    midnight = datetime(year=tomorrow.year, month=tomorrow.month, 
                        day=tomorrow.day, hour=0, minute=0, second=0)
    return (midnight - dt_obj).seconds
nsm_list = []
energy["nsm"] = ""
energy.head()
for index, row in energy.iterrows():
    dt_str = row["date"]
    dt_obj = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
    nsm_list.append(nsm(dt_obj))
# energy["nsm"] = nsm_list
```

```python
from sklearn.base import BaseEstimator, TransformerMixin

def nsm(dt_obj):
    """Get the number of seconds until midnight."""
    tomorrow = dt_obj + timedelta(1)
    midnight = datetime(year=tomorrow.year, month=tomorrow.month, 
                        day=tomorrow.day, hour=0, minute=0, second=0)
    return (midnight - dt_obj).seconds

class CombineAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_nsm=True):
        self.add_nsm = add_nsm
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        for index, row in energy.iterrows():
            dt_str = row["date"]
            dt_obj = datetime.strptime(dt_str, '%Y-%m-%d %H:%M:%S')
            nsm_list.append(nsm(dt_obj))
        print(len(X), len(nsm_list))
        return np.c_[X, nsm_list]
attr_adder = CombineAttributesAdder()
energy_extra_attr = attr_adder.transform(energy.values)
print(energy_extra_attr)
```
